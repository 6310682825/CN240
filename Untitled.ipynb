```python
import cv2
import csv
import os
```

```python
def catch_face(image):
    #อ่านไฟล์สำหรับ classification
    face_cascade = cv2.CascadeClassifier("project240/emotion_detect/haarcascade_frontalface_default.xml")
    #จำแนกใบหน้า scaleFactor's defult = 1.1 , minNeigher's defult = 3 
    scaleFactor = 1.1
    minNeigher = 3
    face_detect = face_cascade.detectMultiScale(image,scaleFactor,minNeigher)
    return face_detect
```

```python
#อ่านภาพ
path = "facial_expressions-master/data/legend.csv"
with open(path) as csvfile:
    reader = csv.reader(csvfile)
    count = 0
    for row in reader:
        if count != 0:
            pic = "facial_expressions-master/images/" + row[1]
            img = cv2.imread(pic)
            #ทำให้ภาพเป็นสีเทา
            gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            good_data = catch_face(gray_img)
            if tuple(good_data) == ():
                new_path = "project240/bad_image"
                cv2.imwrite(os.path.join(new_path, row[1]), gray_img)
                #print("bad",count)
            else:
                print("good")
                #new_path = "project240/good_image"
                #cv2.imwrite(os.path.join(new_path, row[1]), gray_img)
        count += 1
```
